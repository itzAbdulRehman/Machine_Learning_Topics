```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
tips.isnull().sum()
```

```python
sns.countplot(tips['sex'])
```

```python
tips['smoker'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
import matplotlib.pyplot as plt
plt.hist(tips['total_bill'],bins=15)
```

```python
plt.hist(tips['tip'],bins=15)
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
sns.barplot(tips['day'],tips['total_bill'],hue=tips['sex'])
```

```python
sns.heatmap(pd.pivot_table(tips,values='total_bill',index='day',columns='time'))
```

```python
(tips.groupby('day')['total_bill'].mean())
```
